```python
import random
import scipy
from scipy.io import arff
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scikitplot as skplt
import sklearn
from sklearn import preprocessing

from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier

from sklearn.model_selection import *
from sklearn.metrics import *
%config IPCompleter.greedy=True
```

```python
def plot_roc_curve(X_test, y_test, all_clfs, names):
    plt.figure(figsize=(10,8))
    plt.title('Receiver Operating Characteristic')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1.0])
    plt.ylim([0, 1.02])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    
    for clf, name in zip(all_clfs, names):
        probs = clf.predict_proba(X_test)
        preds = probs[:,1]
        fpr, tpr, _ = sklearn.metrics.roc_curve(y_test, preds)
        roc_auc = sklearn.metrics.auc(fpr, tpr)
        # skplt.metrics.plot_roc(y_test, y_pred)
        np.random.seed(abs(hash(name)+hash(clf))%(2**32))
        plt.plot(fpr, tpr, 'b', label = '%s (AUC = %0.2f)' % (name,roc_auc), c=np.random.rand(3,))
        plt.legend(loc = 'lower right')
    plt.savefig('roc')
    plt.show()
```

```python
def plot_pr_curve(X_test, y_test, all_clfs, names):
    plt.figure(figsize=(10,8))
    plt.title('Precision-Recall Curve')
    plt.plot([0, 1], [1, 0],'r--')
    plt.xlim([0, 1.0])
    plt.ylim([0, 1.02])
    plt.ylabel('Precision')
    plt.xlabel('Recall')
    
    for clf, name in zip(all_clfs, names):
        probs = clf.predict_proba(X_test)
        preds = probs[:,1]
        precision, recall, _ = sklearn.metrics.precision_recall_curve(y_test, preds)
        pr_auc = sklearn.metrics.auc(recall, precision)
        # skplt.metrics.plot_roc(y_test, y_pred)
        np.random.seed(abs(hash(name)+hash(clf))%(2**32))
        plt.plot(recall, precision, 'b', label = '%s (AUC = %0.2f)' % (name,pr_auc), c=np.random.rand(3,))
        plt.legend(loc = 'lower left')
    plt.savefig('pr')
    plt.show()
```

```python
X = np.array(df[df.columns[:-1]])
y = np.array(df.Class.astype('int'))
X_train, X_test, y_train, y_test = train_test_split(X, y)
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
def train_KNN():
    print('Training KNN ...')
    knn = KNeighborsClassifier()
    param_distributions = {
        'n_neighbors' : random.sample(range(1,100), 50)
    }
    randcv = sklearn.model_selection.RandomizedSearchCV(knn, param_distributions, n_iter=50, cv=5, n_jobs=-1, iid=False)
    randcv.fit(X_train, y_train)
    return randcv.best_estimator_
```

```python
def train_SVM():
    print('Training SVM ...')
    svm = SVC(kernel='rbf', probability=True)
    param_distributions = {
        'C' : scipy.stats.reciprocal(1.0, 1000.),
        'gamma' : scipy.stats.reciprocal(0.01, 10.),
    }
    randcv = sklearn.model_selection.RandomizedSearchCV(svm, param_distributions, n_iter=30, cv=5, n_jobs=-1, iid=False)
    randcv.fit(X_train, y_train)
    return randcv.best_estimator_
```

```python
def train_DecisionTree():
    print('Training DecisionTree ...')
    tree = DecisionTreeClassifier()
    param_distributions = {
        'max_depth' : random.sample(range(1,100), 50)
    }
    randcv = sklearn.model_selection.RandomizedSearchCV(tree, param_distributions, n_iter=50, cv=5, n_jobs=-1, iid=False)
    randcv.fit(X_train, y_train)
    return randcv.best_estimator_
```

```python
def train_RandomForest():
    print('Training RandomForest ...')
    forest = RandomForestClassifier()
    param_distributions = {
        'max_depth' : random.sample(range(1,100), 50),
        'n_estimators' : random.sample(range(100,200), 50)
    }
    randcv = sklearn.model_selection.RandomizedSearchCV(forest, param_distributions, n_iter=10, cv=5, n_jobs=-1, iid=False)
    randcv.fit(X_train, y_train)
    return randcv.best_estimator_
```

```python
def train_AdaBoost():
    print('Training AdaBoost ...')
    boost = AdaBoostClassifier()
    param_distributions = {
        'learning_rate' : scipy.stats.reciprocal(0.5, 1.5),
        'n_estimators' : random.sample(range(100,200), 50)
    }
    randcv = sklearn.model_selection.RandomizedSearchCV(boost, param_distributions, n_iter=10, cv=5, n_jobs=-1, iid=False)
    randcv.fit(X_train, y_train)
    return randcv.best_estimator_    
```

```python
def train_LogisticRegression():
    print('Training LogisticRegression ...')
    lr = LogisticRegression(solver='liblinear', multi_class='ovr')
    param_distributions = {
        'C' : scipy.stats.reciprocal(0.5, 1000.),
        'max_iter' : random.sample(range(100,200), 50)
    }
    randcv = sklearn.model_selection.RandomizedSearchCV(lr, param_distributions, n_iter=50, cv=5, n_jobs=-1, iid=False)
    randcv.fit(X_train, y_train)
    return randcv.best_estimator_ 
```

```python
def train_GaussianNaiveBayes():
    print('Training GaussianNaiveBayes ...')
    gaussian = GaussianNB()
    param_distributions = {
        'var_smoothing' : scipy.stats.reciprocal(1e-10, 1e-9)
    }
    randcv = sklearn.model_selection.RandomizedSearchCV(gaussian, param_distributions, n_iter=50, cv=5, n_jobs=-1, iid=False)
    randcv.fit(X_train, y_train)
    return randcv.best_estimator_ 
```

```python
def train_NeuralNetwork():
    print('Training NeuralNetwork ...')
    nn = MLPClassifier(solver='lbfgs', hidden_layer_sizes =(20,10,5))
    param_distributions = {
         'learning_rate_init' : scipy.stats.reciprocal(0.001, 0.01),
        'max_iter' : random.sample(range(100,250), 50)
    }
    randcv = sklearn.model_selection.RandomizedSearchCV(nn, param_distributions, n_iter=10, cv=5, n_jobs=-1, iid=False)
    randcv.fit(X_train, y_train)
    return randcv.best_estimator_ 
```

```python
%%time
clf1 = train_KNN()
clf2 = train_SVM()
clf3 = train_DecisionTree()
clf4 = train_RandomForest()
clf5 = train_AdaBoost()
clf6 = train_LogisticRegression()
clf7 = train_GaussianNaiveBayes()
clf8 = train_NeuralNetwork()

all_clfs = [clf1,clf2,clf3,clf4,clf5,clf6,clf7,clf8]
names = ['KNN','SVM','Decision Tree','Random Forest','AdaBoost','Logistic regression','Gaussian Naive Bayes','NeuralNetwork']
plot_roc_curve(X_test,y_test,all_clfs,names)
plot_pr_curve(X_test,y_test,all_clfs,names)
```
